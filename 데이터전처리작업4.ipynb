## 데이터 전처리 작업

### 데이터 변환하기

#### 데이터의 범위를 표준화 또는 정규화를 통해 일치시킬 수 있다. <br><br> 기존 변수를 이용하여 의미 있는 새로운 변수를 생성하거나 변수를 선택할 수 있다.


```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import StandardScaler

import seaborn as sns
```

```python
data = sns.load_dataset('iris')
data = data.drop('species',axis = 1)
print(data.head()) 
```

### 데이터 정규화, 표준화를 통해 범위 맞추기

```python
# 일부 컬럼 scale 하고 싶을 떄

scaler = StandardScaler()
std = scaler.fit_transform(data['sepal_length'].values.reshape(-1,1)) #values.reshape(-1,1) 필요함
print(std)
```

```python
std = pd.Series(std[:,0], name = 'standard') # series화
print(std)
```

```python
#전체 스케일 하고 싶을 때

scale_data = pd.DataFrame(scaler.fit_transform(data))
print(scale_data)
```

```python
data['sepal_length']=scale_data[0]
```
